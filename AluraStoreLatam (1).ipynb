#Visualización de la facturación en cada tienda

```python
ingresos=[
    tienda1['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

colores=rainbow= ['#008fd5', '#fc4f30', '#e5ae38', '#6d904f']

fig, ax = plt.subplots()
barras = ax.bar(tiendas, ingresos, color=colores)

for barra, ingreso in zip(barras, ingresos):
    altura = barra.get_height()
    ax.annotate(f'${ingreso/1e6:.1f} M',
                xy=(barra.get_x() + barra.get_width() / 2, altura),
                xytext=(0, 1),
                textcoords="offset points",
                ha='center', va='bottom', fontsize=10, fontweight='bold'
                )

ax.yaxis.set_ticks([])
ax.set_yticklabels([])

ax.set_title('Ingreso total por cada tienda')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

##Ventas por categorias de cada tienda

```python
tiendas={
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}
```

```python
ventas_categoria = {}
for num_tienda, df in tiendas.items():
    v_categorias=df['Categoría del Producto'].value_counts()
    ventas_categoria[num_tienda] = v_categorias
```

```python
print(ventas_categoria)
```

```python
import matplotlib.pyplot as plt

# figura con 2 filas x 2 columnas
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()  # Convierte a lista para iterar fácilmente

# Iterar sobre las tiendas y los ejes
for ax, (tienda, conteo) in zip(axes, ventas_categoria.items()):
    conteo.plot(kind='bar', ax=ax, color='skyblue')

    # Título y formato
    ax.set_title(f'{tienda} - Categorías del Producto vendidas en cada tienda')
    ax.set_xlabel('')
    ax.set_ylabel('Cantidad vendida')
    ax.set_xticklabels(conteo.index, rotation=45, ha='right')
    ax.grid(axis='y', alpha=0.3)

# Ajustar el espacio para que no se encimen
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda

```python
tiendas={
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

vent_max = {}
vent_min = {}
for num_tienda, df in tiendas.items():
  df['Producto'] = df['Producto'].str.strip()
  vent_max[num_tienda]=df['Producto'].value_counts().head(3)
  vent_min[num_tienda]=df['Producto'].value_counts().tail(3)
for num_tienda in tiendas.keys():
  print(f"El producto más vendido de la {num_tienda} son:\n {vent_max[num_tienda]}\n")
  print(f"El producto menos vendido de la {num_tienda} son:\n {vent_min[num_tienda]}\n")
```

```python
import matplotlib.pyplot as plt

# Crear figura con 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
axes = axes.flatten()  # Convierte matriz en lista para iterar

for ax, num_tienda in zip(axes, tiendas.keys()):
    # Combinar máximos y mínimos de esta tienda
    df_plot = pd.concat([vent_max[num_tienda], vent_min[num_tienda]])

    # Graficar barras
    df_plot.plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')

    # Títulos y etiquetas
    ax.set_title(f'{num_tienda} - Producto más y menos vendido')
    ax.set_xlabel('')
    ax.set_ylabel('Cantidad vendida')
    ax.set_xticklabels(df_plot.index, rotation=45, ha='right')
    ax.grid(axis='y', alpha=0.3)

# Ajustar espacios
plt.tight_layout()
plt.show()
```
